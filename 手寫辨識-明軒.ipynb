```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import os
from keras import utils


# 

flist1 = os.listdir("/content///x_train") # 
x_train = np.empty([len(flist1),50,50])
y_train = np.empty(len(flist1))
x_dict = {"":0, "":1, "":2, "":3}
y_dict = {0:"", 1:"", 2:"", 3:""}
a = 0
# y_train.txt(os.listdir)
with open('/content///y_train.txt', 'w') as f:
  for i in flist1:

    f.write(i[0])
    f.write("\n")

    x_train[a] = cv2.imread(f"/content///x_train/{i}", 0)
    ret, x_train[a] = cv2.threshold(x_train[a], 127, 255, cv2.THRESH_BINARY_INV) # 
    y_train[a] = x_dict[i[0]]
    a = a + 1

# 

flist2 = os.listdir("/content///x_test") # 
x_test = np.empty([len(flist2),50,50])
y_test = np.empty(len(flist2))

a = 0
with open('/content///y_test.txt', 'w') as f:
  for i in flist2:

    f.write(i[0])
    f.write("\n")

    x_test[a] = cv2.imread(f"/content///x_test/{i}", 0)
    ret, x_test[a] = cv2.threshold(x_test[a], 127, 255, cv2.THRESH_BINARY_INV) 
    y_test[a] = x_dict[i[0]]
    a = a + 1

# 
x_train = x_train.reshape(x_train.shape[0],-1)  # 
x_train = x_train.astype('float32')/255      # 
y_train = y_train.astype(np.float32)

# 
x_test = x_test.reshape(x_test.shape[0],-1)    # 
x_test = x_test.astype('float32')/255       # 
y_test = y_test.astype(np.float32)

knn=cv2.ml.KNearest_create()       #  KNN 
knn.setDefaultK(16)            # 
knn.setIsClassifier(True)

print('training...')
knn.train(x_train, cv2.ml.ROW_SAMPLE, y_train)  # 
knn.save('tch_knn.xml')       # 
print('ok')

print('testing...')
test_pre = knn.predict(x_test)    # 
test_ret = test_pre[1]
test_ret = test_ret.reshape(-1,)
test_sum = (test_ret == y_test)
acc = test_sum.mean()        # 
print(acc)
```
